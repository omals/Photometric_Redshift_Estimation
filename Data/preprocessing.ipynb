```python
### This script reads three CSV files, counts the number of rows in each,
### and finds common rows across all three datasets.

import pandas as pd

# Load the three CSV files
df1 = pd.read_csv('D:\\Surprise\\withoutbackground\\hira\\SDSS\\SDSS1_Omal.csv')
df2 = pd.read_csv('D:\\Surprise\\withoutbackground\\hira\\SDSS\\SDSS2_Omal.csv')
df3 = pd.read_csv('D:\\Surprise\\withoutbackground\\hira\\SDSS\\SDSS3_Omal.csv')

# Ensure all DataFrames have the same column order
df2 = df2[df1.columns]
df3 = df3[df1.columns]

# Count total rows excluding headers
rows_df1 = df1.shape[0]
rows_df2 = df2.shape[0]
rows_df3 = df3.shape[0]
total_rows = rows_df1 + rows_df2 + rows_df3

print(f"Total rows in SDSS1: {rows_df1}")
print(f"Total rows in SDSS2: {rows_df2}")
print(f"Total rows in SDSS3: {rows_df3}")
print(f"Total number of rows across all datasets: {total_rows}")

# Find common rows across all three
common_rows = pd.merge(df1, df2, how='inner')
common_rows = pd.merge(common_rows, df3, how='inner')

# Print results
if not common_rows.empty:
    print(f"\nFound {len(common_rows)} common rows across all 3 datasets:")
    print(common_rows)
else:
    print("\nNo common rows found in all three datasets.")
```

```python
#Combine the three CSV files and find common row

# Combine all into one DataFrame
combined_df = pd.concat([df1, df2, df3], ignore_index=True)

# Save to a new CSV file
combined_df.to_csv('D:\\Surprise\\withoutbackground\\hira\\SDSS\\SDSS_combined.csv', index=False)

print(f"Combined CSV created with {len(combined_df)} rows.")
```

```python
import pandas as pd
combined_df = pd.read_csv('D:\\Surprise\\withoutbackground\\hira\\SDSS\\SDSS_combined.csv')
print(combined_df.head())
```

```python
for column in combined_df.columns:
    if pd.api.types.is_numeric_dtype(combined_df[column]):
        min_val = combined_df[column].min()
        max_val = combined_df[column].max()
        print(f"{column}: min = {min_val}, max = {max_val}")
    else:
        print(f"{column}: non-numeric column (e.g., string or category)")
```

```python

print(combined_df.shape)
# 1. Remove rows where 'redshift' is missing or empty
combined_df = combined_df[combined_df['redshift'].notna()]  # Drop rows with NaN redshift
combined_df = combined_df[combined_df['redshift'] != '']    # Drop empty string values
print(combined_df.shape)
# 2. Remove non-numeric entries from all columns
for col in combined_df.columns:
    combined_df[col] = pd.to_numeric(combined_df[col], errors='coerce')

# Drop rows with any NaN values caused by non-numeric conversion
combined_df = combined_df.dropna()

# 3. Drop astrometry & metadata columns
columns_to_drop = [
    'ra', 'dec', 'raErr', 'decErr', 'cx', 'cy', 'cz', 'l', 'b',
    'objID', 'run', 'rerun', 'camcol', 'field',
    'type', 'mode', 'nChild', 'clean', 'probPSF', 'skyVersion', "redshiftWarning", 'loadVersion','htmID', 'fieldID', 'parentID','specObjID'
]

combined_df = combined_df.drop(columns=[col for col in columns_to_drop if col in combined_df.columns], errors='ignore')
print(combined_df.shape)
# Save the cleaned dataset
# combined_df.to_csv("D:\\Surprise\\withoutbackground\\hira\\SDSS\\sdss_cleaned.csv", index=False)

print("Dataset cleaned")
```

```python
import seaborn as sns

combined_df = pd.read_csv('D:\\Surprise\\withoutbackground\\hira\\SDSS\\sdss_cleaned.csv')
# Keep only numeric features
df_numeric = combined_df.select_dtypes(include='number')

# Compute correlation matrix
correlation_matrix = df_numeric.corr()

# Sort correlations with redshift
redshift_corr = correlation_matrix['redshift'].sort_values(ascending=False)

# Display the top correlated features
print("Top features positively correlated with redshift:")
print(redshift_corr)  # Exclude self-correlation at index 0

print("\nTop features negatively correlated with redshift:")
print(redshift_corr[-5:])

# Plot correlation heatmap with redshift
plt.figure(figsize=(10, 8))
sns.heatmap(df_numeric.corr()[['redshift']].sort_values(by='redshift', ascending=False),
            annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation with Redshift')
plt.show()
```
